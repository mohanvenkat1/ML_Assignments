1. Basic Line Plot
o Plot a simple line graph of ùë¶ = ùë•2 for ùë• values ranging from -10 to 10.
o Label the x-axis as "X Values" and the y-axis as "Squared Values."
o Add a title to the graph.

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(-10,10)
y=x*x
plt.title('Squared Values')
plt.xlabel('X Values')
plt.ylabel('Squared Values')
plt.plot(x,y,color='red')
plt.show()
```

8. Comparison of Matplotlib and Seaborn
o Generate a pair of plots (bar charts or line plots) using the same data in
both Matplotlib and Seaborn.
o Highlight the differences in the code, syntax, and aesthetics of the two
libraries.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
x = np.random.randint(1, 10, 10)
y = np.random.randint(1, 100, 10)
plt.bar(x, y)
plt.title("Matplotlib Bar Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
sns.barplot(x=x, y=y)
plt.title("Seaborn Bar Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

```

9. Pair Plot (Seaborn)
o Manually create a small dataset with columns such as 'Age', 'Height',
'Weight' for 10 individuals and visualize the relationships using a pair plot.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    'Age': [25, 30, 35, 40, 28, 32, 45, 50, 38, 42],
    'Height': [5.5, 5.7, 6.0, 5.8, 5.6, 5.9, 6.1, 5.4, 5.8, 6.0],
    'Weight': [60, 65, 70, 75, 62, 68, 80, 55, 72, 78]
}
df = pd.DataFrame(data)
sns.pairplot(df, diag_kind='kde', corner=True)
plt.show()
```

10.
Save and Style Plots
‚Ä¢ Plot any graph of your choice and save it in both .png and .pdf formats.
‚Ä¢ Apply at least one built-in style (e.g., 'ggplot' or 'seaborn-darkgrid') and explain
how the style impacts the visualization.

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Sine Wave', color='blue', linewidth=2)
plt.title('Styled Sine Wave Plot', fontsize=16)
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.legend()
plt.show()
```
